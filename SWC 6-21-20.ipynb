```python
first = pd.read_csv('data/gapminder_all.csv', index_col='country')
```

```python
first['continent'] == 'Americas'
```

One of the columns in gapminder is continent. So we're asking it with above code to "give me true when continent = Americas"

Note: == is an piece of code that returns "True" when two things are equal. If a = 1, and you type "a == 1", then it will return "True."

```python
second = first[first['continent'] == 'Americas']
```

The above code indexes into the countries defined as "First" and define only counties that return True to the statement.

third = second.drop('Puerto Rico')

The above code takes 'second' and simply removes the puerto rico row.

```python
second
```

```python
third
```

```python
fourth = third.drop('continent', axis = 1)
```

The above code drops the COLUMN "continent" because you've defined that the argument switches its axis (from default, rows) to columns (axis = 1)

```python
fourth.to_csv('result.csv')
```
