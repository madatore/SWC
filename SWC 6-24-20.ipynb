```python
import pandas as pd
```

```python
data = pd.read_csv('data/gapminder_gdp_oceania.csv', index_col='country')

years = data.columns.str.strip('gdpPercap_')

data.columns = years.astype(int)

data.loc['Australia'].plot()
```

In order to plot things, we can't plot strings, we can only plot numbers against numbers.

You can run "strip" on bits of a string; and strip off, for example "gdpPercap_" leaving just the years.

Then you can recategorize them as integers with "data.columns = years.astype(int)"

How did it know to plot years as the x axis and GDP as the y axis? By default, this plot knows to plot colums as the x-axis.

```python
data.T
```

^^a transpose of our data

```python
data.T.plot()
plt.ylabel('GDP_per_capita')
```
